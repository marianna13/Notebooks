```
import numpy as np
def f1(x,y):
  return 2*x
def f(x,y,m,n):
  return f1(x,y)*np.sin(m*np.pi/b*x)*np.sin(n*np.pi/d*x)
def s(a,b,c,d,m,n):
  s = (16*f((b+a)/2,(c+d)/2,m,n)+4*f((b+a)/2,d,m,n)+4*f((b+a)/2,c,m,n)+4*f((b+a)/2,d,m,n)+4*f(b,(d+c)/2,m,n)+4*f(a,(d+c)/2,m,n)+f(b,d,m,n)+f(a,d,m,n)+f(a,c,m,n)+f(b,c,m,n))*(b-a)*(d-c)/36
  return s
def A(m,n, *cache):
  a,b,c, d= cache
  return 4/(b*d)*s(a,b,c,d,m,n)
a=0
b= 10
c=0
d=100
X = np.linspace(1,2,10)
Y = np.linspace(1,2,10)


X,Y = np.meshgrid(X,Y)
C= 1.22e-3
def u(a,b,c,d,C,t,X,Y):
  u=0
  for m in range(1,10):
    for n in range(1,10):
      mu = m*np.pi/b
      nu = n*np.pi/d
      lmn = C*np.sqrt(mu**2*nu**2)
      u+=(A(m,n,a,b,c,d)*np.sin(mu*X)*np.sin(nu*Y)*np.exp(-lmn**2*t))
  return u
for t in range(1,10):
  ut = u(a,b,c,d,C,t,X,Y)
print(type(ut))
```

```
import plotly.graph_objects as go
for t in range(1,10):
  ut = u(a,b,c,d,C,t,X,Y)
  fig = go.Figure(data=[go.Surface(z=ut, x=X, y=Y)])
  fig.update_layout(title='', autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))
  fig.show()
```
